```python
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
path = "/home-local/Kernel_Conversion/AnatomyConstrainedMultipathGAN/vanillacyclegan_anatomy_B50ftoB30f/web/images"
real_b50f = Image.open(os.path.join(path, "epoch004_real_A.png")).convert('L')
fake_b30f = Image.open(os.path.join(path, "epoch004_fake_B.png")).convert('L')
reconstructed_b50f = Image.open(os.path.join(path, "epoch004_rec_A.png")).convert('L')
```

```python
b50f = np.rot90(np.array(real_b50f).astype(np.float32))
b30f = np.rot90(np.array(fake_b30f).astype(np.float32))
reconstructed_b50f = np.rot90(np.array(reconstructed_b50f).astype(np.float32))
diff = b50f - b30f
diffrot = np.rot90(diff)

#Convert from 0 to 255 to hounsfield units 
b50hu = (b50f/255) * (3072 - (-1024)) + (-1024)
b30hu = (b30f/255) * (3072 - (-1024)) + (-1024)
reconstructed_b50hu = (reconstructed_b50f/255) * (3072 - (-1024)) + (-1024)
diff = b50hu - b30hu 

```

```python
fig, ax = plt.subplots(1, 4, figsize=(15, 5))
ax[0].imshow(b50f, cmap='gray', vmin=-1024, vmax = 600)
ax[0].set_title("Real B50f")
ax[1].imshow(b30f, cmap='gray', vmin=-150, vmax = 150)
ax[1].set_title("Fake B30f")
ax[2].imshow(reconstructed_b50f, cmap='gray', vmin=-150, vmax = 150)
ax[2].set_title("Reconstructed B50f")
ax[3].imshow(diff, cmap='gray', vmin=-150, vmax = 150)
ax[3].set_title("Difference between B50f and B30f")
plt.show()
```
